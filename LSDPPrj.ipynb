```python
#######Query-2########
#######Days Of Month That See Maximum Arrival Delays########
ds = pd.read_csv("Query2.csv")
ds.columns=['MonthDay','ARRIVAL_DELAY']
```

```python
ds.head(5)
```

```python
from ggplot import *
```

```python
g2=ggplot(aes(x='MonthDay', y='ARRIVAL_DELAY'), data=ds) +geom_line()+geom_point()
g2.show()
```

```python
g3=ggplot(aes(x='MonthDay', y='ARRIVAL_DELAY'), data=ds) +geom_histogram()+geom_jitter()
g3.show()
```

```python
#####Inference Drawn-11th of any month sees the most number of arrival delays######
```

```python
#############################Query-9##########################
#####Busiest Airports as per the days of the week########
```

```python
d9=pd.read_csv("Query9.csv")
d9.columns=['WeekDay','Airport','FlightCount']
d9.head(5)
```

```python
d9.plot.bar(x = 'Airport', y = ['WeekDay', 'FlightCount'], rot =0)
plt.show()
```

```python
fig, ax = plt.subplots()
d9.plot.bar(x = 'Airport', y = ['WeekDay', 'FlightCount'], rot = 40, ax = ax)
for p in ax.patches: 
    ax.annotate(np.round(p.get_height(),decimals=2), (p.get_x()+p.get_width()/2., p.get_height()))
plt.show()
```
